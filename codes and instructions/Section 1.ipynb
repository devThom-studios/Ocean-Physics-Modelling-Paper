```python
# import MITgcmutils
import MITgcmutils
# define path to ctrl simulation
ctrl = '/home/jovyan/MITgcm/my__exp/ctrl/'
# open grid* files and load global X coordinate of cell center
nc = MITgcmutils.mnc_files(ctrl + 'mnc*/grid.t*.nc')
xc = nc.variables['XC'][:]
yc = nc.variables['YC'][:]
nc.close()
# open Diag* files and load global potential temperature
ncv = MITgcmutils.mnc_files(ctrl + 'mnc*/Diag.*.nc')
ncw = MITgcmutils.mnc_files(ctrl + 'mnc*/Sdiag_MLD.*.nc') 
ptemp = ncv.variables['THETA'][:]
salt=ncv.variables['SALT'][:]
mld=ncw.variables['MXLDEPTH'][:]
ncv.close()
ncw.close()
```

```python
from matplotlib import pyplot as plt
fig, ax=plt.subplots(3,1, figsize=(10,15))
a=ax[0].pcolormesh(xc, yc[:], ptemp[9, 0, ...])
plt.colorbar(a)
b=ax[1].pcolormesh(xc, yc, salt[9, 0, ...],  vmin=32, vmax=37)
plt.colorbar(b)
c=ax[2].pcolormesh(xc, yc, mld[9, 0, ...])
plt.colorbar(c);
```
